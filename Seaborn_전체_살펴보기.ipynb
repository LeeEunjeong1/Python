## seaborn - simple example

```

import matplotlib.pyplot as plt
import seaborn as sns
```

```
tips = sns.load_dataset("tips")
sns.set_style("whitegrid")
g = sns.lmplot(x="tip", y="total_bill", data=tips, aspect=2)
g = (g.set_axis_labels("Tip","Total bill(USD)").set(xlim=(0,10),ylim=(0,100)))
plt.title("title")
plt.show(g)
```

## 1. Data

```

import pandas as pd
import numpy as np
uniform_data = np.random.rand(10, 12)
data = pd.DataFrame({'x':np.arange(1,101), 'y':np.random.normal(0,4,100)})
```

```
# s built-in data sets:
titanic = sns.load_dataset("titanic")
iris = sns.load_dataset("iris")
```

## 2. Figure Aesthetics

```
f, ax = plt.subplots(2, 1, figsize=(5,6))
```

```
f, ax = plt.subplots(figsize=(16,8)) 
```

```
# Seaborn styles
sns.set() # (Re)set the seaborn default
sns.set_style("whitegrid") # Set the matplotlib parameters
sns.set_style("ticks", {"xtick.major.size":8, "ytick.major.size":8})
sns.axes_style("whitegrid") 
```

```
# Context Functions
sns.set_context("talk") # Set context to "talk"
sns.set_context("notebook", font_scale=1.5, rc={"lines.linewidth":2.5}) # Set context to "notebook"
```

```
# Color Palette
sns.set_palette("husl", 3) # Define the color palette
sns.color_palette("husl") # Use with with to temporarily set palette
flatui = ["#9b59b6","#3498db","#95a5a6","#e74c3c","#34495e","#2ecc71"]
sns.set_palette(flatui) # Set your own color palette
```

## Plotting With Seaborn

### Axis Grids

```
g = sns.FacetGrid(titanic, col="survived", row="sex")
g = g.map(plt.hist,"age")
```

```
g = sns.factorplot(x="pclass", y="survived", hue="sex", data=titanic)
```

```
g = sns.lmplot(x="sepal_width", y="sepal_length", hue="species", data=iris)
```

```
h = sns.PairGrid(iris)
h = h.map(plt.scatter)
```

```
sns.pairplot(iris, hue='species')
```

```
i = sns.JointGrid(x="x", y="y", data=data)
i = i.plot(sns.regplot, sns.distplot)
```

```
sns.jointplot("sepal_length", "sepal_width", data=iris, kind='kde')
```

## Categorical Plots

```
# Scatterplot
sns.stripplot(x="species", y="petal_length", data=iris)
```

```
# Swarmplot (Scatterplot)
sns.swarmplot(x="species", y="petal_length", data=iris) # non-overlapping
```

```
# Bar Chart
sns.barplot(x="sex", y="survived", hue="class", data=titanic)
```

```
# Count Plot
sns.countplot(x="deck", data=titanic, palette="Greens_d")
```

```
# Point Plot
sns.pointplot(x="class", y="survived", hue="sex", data=titanic, 
	palette={"male":"g", "female":"m"}, markers=["^","o"], linestyles=["-","--"])
```

```
# Boxplot
sns.boxplot(x="alive", y="age", hue="adult_male", data=titanic)
```

```
sns.boxplot(data=iris, orient="h") # wide-form
```

```
# Violinplot
sns.violinplot(x="pclass", y="age", hue="survived", data=titanic)
```

```
sns.violinplot(x="pclass", y="age", hue="survived", data=titanic, split=True)
```

```
f,ax = plt.subplots(1,2,figsize=(16,8))

sns.violinplot("pclass","age", hue="survived", data=titanic, scale='count', split=True, ax=ax[0])
ax[0].set_title('pclass and age vs survived')
ax[0].set_yticks(range(0,110,10))

sns.violinplot("sex","age", hue="survived", data=titanic, scale='count', split=True,ax=ax[1])
ax[1].set_title('sex and age vs survived')
ax[1].set_yticks(range(0,110,10))

plt.show()
```

### Regression Plots

```
sns.regplot(x="sepal_width", y="sepal_length", data=iris)
```

### Distribution Plots

```
sns.distplot(data.y, kde=True, color="b")
```

```
sns.kdeplot(titanic[titanic['survived'] == 1]['age'])
sns.kdeplot(titanic[titanic['survived'] == 0]['age'])

plt.legend(['survived=1', 'survived=0'])
```

### Matrix Plots

```
sns.heatmap(uniform_data,vmin=0,vmax=1) 
```

## Further Customizations

### Axisgrid Objects

```
g = sns.lmplot(x="tip", y="total_bill", data=tips, aspect=2)

g.despine(left=True)
g.set_ylabels("Survived")
g.set_xticklabels(rotation=45) 
g.set_axis_labels("Survived", "Sex")
```

```
g = sns.pairplot(iris, hue='species')
g.set(xlim=(0,10), ylim=(2,8), xticks=[0,5,10], yticks=range(2,8))
```
